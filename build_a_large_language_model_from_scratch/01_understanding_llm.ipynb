#### Stages of building LLM

- pretuning  (trained on diverse dataset) > finetuning (trained on specialized dataset)
- pretuned model - base model
- 2 types of finetuning
    - instruction finetuning
    - finetuning for classifciation
- most llm are transformer based (deep learning)
    - transformers - encoder and decoder
    - input > preproccessing > encode > embeddings > decoder > output layer
    - BERT (bidirectional encoder representations from tranformers)
    - GPT (generative pretrained transformers)
    - BERT (masked word prediction, focused on encoder) / GPT (generative for next word, focues on decoder)



- Pretraining LLMs requires access to significant resources and is very expensive. For example, the GPT-3 pretraining cost is estimated to be $4.6 million in terms of cloud computing credits
-  LLMs can be finetuned on specific tasks with relatively smaller datasets, reducing the computational
resources needed and improving performance on the specific task.
- It's interesting to note that although the original transformer model was explicitly designed for language translation, GPT models—despite their larger yet simpler architecture aimed at next-word prediction—are also capable of performing translation tasks
- in pretraing phase, LLMs leverage self-supervised learning, where the model generates its own labels from the input data
