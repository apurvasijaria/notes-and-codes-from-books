````python
## Creating the test set
def split_train_test(data, test_ratio):
    """Splits a dataset into train/test using a `test_ratio`.
    
    # Arguments
        data (pd.DataFrame): the full dataset.
        test_ratio (float): the percentage taken from the original dataset as a test set.
    
    # Returns
        train (pd.DataFrame): the training set.
        test (pd.DataFrame): the test set.
    """
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
````

````python
## Another Implementation of same datasplit usign id
from zlib import crc32

def test_set_check(identifier, test_ratio):
    '''
    function to split test and train based on id and test_ratio
    
    #Arguments:
    identifier (list): unique and immutable identifier
    test_ratio (float): ratio of test by total data set
    
    #Returns:
    list: returns the list of ids in test set 
    '''
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32


def split_train_test_by_id(data, test_ratio, id_column):
    '''
    function to split train and test based on id and test_ratio
    #Arguments:
    data (pd.DataFrame): full data set
    test_ratio (float): ratio of test by total data set
    id_column (character): column name of the identifier column in full dataset 
    
    #Returns:
    train_set (pd.DataFrame): train set 
    test_set (pd.DataFrame): test set 
    
    '''
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
````

### Scikit-learn design

Main design principles:

##### Consistency

- **Estimators**: object that can predict some paramters based on a dataset. performed by ```fit()``` method, with dataset as a parameter
- **Transformers**: some estimators can transform the dataset, they are called transformers. performed by ```transform()``` method, with the dataset as a parameter. all transformers have a method ```fit_transform()``` which can do both estimation and transformation in single method, and is also faster to use. 
- **Predictors**: some estimators are capable of making predictions, with dataset as a paramter. predictors have a method ```predict()``` that takes a new dataset. it also has ```score()``` method, with measures the quality of the predictions given the predictions and the actual target values.

##### Inspection

the estimator's hyperparamters are accessible directly with the method and underscore suffix, eg: ```imputer.statistics_```

##### Non proliferation of classes

- datasets are represented as numpy or scipy sparse matrices, instead of homemade classes
- hyperparameters are regular python strings or numbers

##### Composition

existing building blocks are reused as much as possible

##### Sensible defaults

scikit-learn provides a default value for most paramters, making it easy for baseline working systems

#### Custom Transformers

TransformerMixin allows adding the hyperparameters and new feature of which we are 100% sure to add. New combinations can be automated using this. 

````python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): 
        # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        return self # nothing else to do
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]


attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
````

#### Transformation Pipelines

scikit-learn provides pipeline class to sequence the data transformation/preprocessing steps


````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy = "median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler',StandardScaler()),
])
````

````python
housing_num_tr = num_pipeline.fit_transform(housing_num)
````

To handle cateogrical and numerical variable together, using ColumnTransformer on each column

_https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html_

_https://machinelearningmastery.com/columntransformer-for-numerical-and-categorical-data/_

````python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline,num_attribs),
    ("cat",OneHotEncoder(),cat_attribs),    
])
````

````python
housing_prepared = full_pipeline.fit_transform(housing)
````

Note that the ```OneHotEncoder```  returns a sparse matrix, while the ```num_pipeline``` returns a  dense  matrix.  When  there  is  such  a  mix  of  sparse  and  dense  matrices,  the ```ColumnTransformer```  estimates  the  density  of  the  final  matrix  (i.e.,  the  ratio  of  non-zero cells), and it returns a sparse matrix if the density is lower than a given threshold (by
default,  sparse_threshold=0.3 ).  In  this  example,  it  returns  a  dense  matrix.  And that’s it! We have a preprocessing pipeline that takes the full housing data and applies the appropriate transformations to each column.

Instead  of  a  transformer,  you  can  specify  the  string  "drop"  if  you want  the  columns  to  be  dropped.  Or  you  can  specify  "pass through" if you want the columns to be left untouched. By default,the  remaining  columns  (i.e.,  the  ones  that  were  not  listed)  will  be dropped,  but  you  can  set  the  remainder  hyperparameter  to  any transformer (or to "passthrough") if you want these columns to be handled differently

In some cases, the generalization error maybe not enough for a production model. In these cases, it's better to test a _95% Confidence Interval_ for generlization error using ```scipy.stats.t.interval()```


````python
from scipy import stats

confidence_thresh = [0.90,0.95,0.99]
squared_errors = (final_predictions - y_test) ** 2

for confidence in confidence_thresh:
    print('Confidence Interval: ',confidence)
    print(np.sqrt(stats.t.interval(confidence, len(squared_errors)-1,
                        loc = squared_errors.mean(),
                        scale= stats.sem(squared_errors))))
    print('-------------------')
````

**3. Try adding a transformer in the preparation pipeline to select only the most important attributes.**

````python
def rank_features(feature_list,k):
    '''
    function to select k best features from a list, returns the list of 
    indices of the best k features
    
    #Arguments:
    feature_list (list): feature_importance list
    k (int): number of top features requested
    
    #Returns:
    Indices of the best k features
    '''
    return np.argpartition(feature_list, -k)[-k:]
````

````python
from sklearn.base import BaseEstimator, TransformerMixin

class SelectkBestFeatures(BaseEstimator, TransformerMixin):
    '''
    Custom transformer to select top k features from the list of features
    
    #Arguments:
    k (int): number of top features requested
    feature_importance(list): feature_importance list
    
    #Returns:
    The list of best features
    
    '''
    def __init__(self, feature_imp, k ): 
        self.feature_imp = feature_imp
        self.k = k
    
    def fit(self, X, y=None):
        self.feature_rank_ = np.argpartition(np.array(self.feature_imp), -self.k)[-self.k:]
        return self 
    
    def transform(self, X):
        return X[:, self.feature_rank_]

````
